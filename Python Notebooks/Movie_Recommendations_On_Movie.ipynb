```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from ast import literal_eval
```

```python
data_movies = pd.read_csv('C:\\Users\\ujwal\\Downloads\\movies.csv',na_filter=False)

data_credits = pd.read_csv('C:\\Users\\ujwal\\Downloads\\credits.csv')
```

```python
#Merge credits and movie datasets on movie_id
data_movies = data_movies.merge(data_credits,on='movie_id')
```

```python
data_movies = data_movies.drop(columns=['title_x','title_y', 'tagline', 'status', 'homepage', 
                                        'keywords','crew','vote_count', 'vote_average',
                                       'tagline', 'spoken_languages', 'runtime',
                                       'popularity', 'production_companies', 'budget',
                                       'production_countries','release_date', 'revenue',
                                         'original_language'])

data_movies.head(5)
```

```python
#Rename Overview Column to Plot
data_movies.rename(columns = {'overview':'plot'}, inplace = True)
```

```python
#Convert cast and genre column to proper format
features = ['cast', 'genres']
for feature in features:
    data_movies[feature] = data_movies[feature].apply(literal_eval)
```

```python
#Pick only top 3 genre and cast from respective columns
def get_list(meta_data):
    if isinstance(meta_data, list):
        names = [col['name'] for col in meta_data]
        if len(names) > 3:
            names = names[:3]
        return names
    return []

features = ['cast','genres']
for feature in features:
    data_movies[feature] = data_movies[feature].apply(get_list)
```

```python
data_movies.to_csv (r'C:\Users\ujwal\Downloads\Recommendations_DataSet_Movie.csv', index = False, header=True)
```
