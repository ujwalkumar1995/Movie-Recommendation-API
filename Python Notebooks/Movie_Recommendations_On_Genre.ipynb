```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from ast import literal_eval
```

```python
data_movies = pd.read_csv('C:\\Users\\ujwal\\Downloads\\movies.csv',na_filter=False)
```

```python
#Delete Unnecessary Columns
data_movies = data_movies.drop(columns=['title', 'tagline', 'status', 'homepage', 
                                       'popularity', 'production_companies', 'budget',
                                       'production_countries','keywords','overview','runtime','spoken_languages','release_date', 'revenue',
                                         'original_language'])
```

```python
data_movies.head(5)
```

```python
#Convert Genre Column to Proper Format
features = ['genres']
for feature in features:
    data_movies[feature] = data_movies[feature].apply(literal_eval)
```

```python
#Pick only top 3 genres
def get_list(meta_data):
    if isinstance(meta_data, list):
        names = [col['name'] for col in meta_data]
        if len(names) > 3:
            names = names[:3]
        return names
    return []

features = ['genres']
for feature in features:
    data_movies[feature] = data_movies[feature].apply(get_list)
```

```python
#Apply one hot encoding to the genre column
data_movies = data_movies.drop('genres', 1).join(
    pd.get_dummies(
        pd.DataFrame(data_movies.genres.tolist()).stack()
    ).astype(int).sum(level=0)
)
```

```python
data_movies.head(5)
```

```python
#Drop rows with NA
data_movies = data_movies.dropna()
```
